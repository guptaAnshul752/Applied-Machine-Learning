## 68-95-99.7 Rule Verification

```python
# Taking normal_10000 for reference
```

```python
# One Standard Deviation Right
one_std_right = mean + std

# One Standard Deviation Left
one_std_left = mean - std

# Two Standard Deviation Right
two_std_right = mean + 2*std

# Two Standard Deviation Left
two_std_left = mean - 2*std

# Three Standard Deviation Right
three_std_right = mean + 3*std

# Three Standard Deviation Left
three_std_left = mean - 3*std
```

```python
# Plotting Normal_10000 Distribution

plt.figure(figsize = (20,10))

sns.set_style('darkgrid')

sns.distplot(normal_10000)

plt.axvline(mean,color = 'coral',label = 'Mean')

plt.axvline(one_std_right,color = 'red',label = 'Mean + 1SD')
plt.axvline(one_std_left,color = 'red',label = 'Mean - 1SD')

plt.axvline(two_std_right,color = 'yellow',label = 'Mean + 2SD')
plt.axvline(two_std_left,color = 'yellow',label = 'Mean - 2SD')

plt.axvline(three_std_right,color = 'green',label = 'Mean + 3SD')
plt.axvline(three_std_left,color = 'green',label = 'Mean - 3SD')

plt.title("Normal Distribution")

plt.legend()
plt.show()
```

```python
normal_10000 < one_std_right
```

```python
# 68% Rule

((normal_10000 < one_std_right) & (normal_10000 > one_std_left)).sum() / len(normal_10000) * 100
```

```python
# 95% Rule

((normal_10000 < two_std_right) & (normal_10000 > two_std_left)).sum() / len(normal_10000) * 100
```

```python
# 99.7% Rule

((normal_10000 < three_std_right) & (normal_10000 > three_std_left)).sum() / len(normal_10000) * 100
```

## Checking Skewness & Kurtosis

#### Skewness

In statistics, skewness is a measure of the asymmetry of the probability distribution of a random variable about its mean. In other words, skewness tells you the amount and direction of skew (departure from horizontal symmetry). The skewness value can be positive or negative, or even undefined. If skewness is 0, the data are perfectly symmetrical, although it is quite unlikely for real-world data. As a general rule of thumb:

- If skewness is less than -1 or greater than 1, the distribution is highly skewed.
- If skewness is between -1 and -0.5 or between 0.5 and 1, the distribution is moderately skewed.
- <b>If skewness is between -0.5 and 0.5, the distribution is approximately symmetric.</b>

#### Kurtosis

Kurtosis tells you the height and sharpness of the central peak, relative to that of a standard bell curve.
